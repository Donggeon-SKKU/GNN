#### Degree

degree of Undirected Graph
+ the degree of vertex in a graph is the number of deges connected to it
+ denote the degree of vertex $i$ by $d_i$
+ for an undirected graph of $n$ vertices $$d_i=\sum_{j=1}^nA_{ij}$$
+ Degree matrix $D$ of adjacent matrix $A$ $$D=diag\{d_1,d_2,\dots\}$$
+ example



$$A=\begin{bmatrix}
0 & 1 & 1 & 1 \\
1 & 0 & 0 & 0 \\
1 & 0 & 0 & 1 \\
1 & 0 & 1 & 0 \\
\end{bmatrix} \rightarrow 

D=\begin{bmatrix}
3 & 0 & 0 & 0 \\
0 & 1 & 0 & 0 \\
0 & 0 & 2 & 0 \\
0 & 0 & 0 & 2 \\
\end{bmatrix}$$

#### Self Connected Edge

$$A=\begin{bmatrix}

\end{bmatrix} \rightarrow 

A+I = \begin{bmatrix}

#### 2. Graph Convolution Network(GCN)

##### 2.1 Convolution
+ In previous CNN lecture CNN has two charateristics, preserving the spatial structure and weight sharing
+ To apply convolution in graph network, graph also has to conpensate that characteristics too

__Convolution Layer__
+ In CNN, convolution layer preserve the spatial structure of input
+ It convolve over all spatial locations
    + Extract features for each convoltion layer
    

__Weight Sharing__
+ Reduce the number of parameters by weight sharing
+ within the same layer, the same filter will be used through image


##### 2.2. Connection beteween CNN and GCN

+ GCNs perform similar operations where the model learns the features by inspecting neighboring nodes
+ The major difference between CNNs and GCNs is that CNNs are specially built to operate on regular (Euclidean) structured data, while GCNs operate for the graph date that the number of nodes connections cary and the nodes are unordered (irregular on non-Euclidean structured data)


##### 2.3. Basics of GCN

+ similar to CNN, GCN updates each node with their adjacent nodes
+ Unlike CNN, each node of GCN has different number of adjacent nodes
    + Indicate adjacent nodes of each node by adjacency matrix $A$
+ basic process (or terminology) of GCN
    + Message: information passed by neighboring nodes to the central node
    + Aggregate: collect information from neighboring nodes
    + Update: embedding update by combining information from neighboring nodes and from itself


$$h_u^{(k+1)}=UPDATE(AGGREGATE({h_v^{(k)},\forall{v}\in\mathcal{N}(u)}))$$
1) Message Aggregation from Local Neighborhood
$$AGGREGATE({h_v^{(k)},\forall{v}\in\mathcal{N}(u)}) \\ \Rightarrow AH^{(k)}$$


2) Update

Adding a non-linear function: $k^{th}$ layer
$$H^{(k+1)} = f(A,H^{(k)}) \\ = \sigma(AH^{(k)}W)$$
$$h_v^{(k+1)}=\sigma(AH^{(k)}W^{k}_{neigh})$$

+ $h_1^{k}$:feature matrix of first node in $k^{th}$ layer
+ $W^{(k)}$: weight of $k^{th} layer
    + Whight sharing: share same weight for each layer
        + In the same layer, each node is updated similarly, so it shares the same weight
        + whight sharing enhanced computing complexity and time

        


##### 2.4. Further Improvement for GCN

1) Message Passing with Self-Loops
+ As a simplification of the neural message passing approach, it is common to add self-loops to the input graph and omit the explicit update step

$$\begin{aligned}
h_i^{(k+1)} & =\operatorname{UPDATE}\left(h_u^{(k)}, \operatorname{AGGREGATE}\left(\left\{h_v^{(k)}, \forall v \in \mathcal{N}(u)\right\}\right)\right) \\
& =\operatorname{UPDATE}\left(\operatorname{AGGREGATE}\left(\left\{h_v^{(k)}, \forall v \in \mathcal{N}(u) \cup\{u\}\right\}\right)\right) \\
H^{(k+1)} & =\sigma\left((A+I) H^{(k)} W^{(k)}\right)
\end{aligned}$$




2) Neighborhood Normalization
+ The most basic neighborhood aggregation operation simply takes the sum of the neighbor embedding.
+ One issue with this approach is that it can be unstable and highly sensitive to node degrees.
+ One silution to this problem is to simply normalize the aggretion operation based upon the degrees of the nodes involbed
+ The simplest approach is to just take a weighted average rather than sum

$$\begin{aligned}
\bar{A} & =D^{-1 / 2} A D^{-1 / 2}+I \\
& \approx \tilde{D}^{-1 / 2}(A+I) \tilde{D}^{-1 / 2} \quad \text { where } \tilde{D} \text { is the degree matrix of } A+I
\end{aligned}$$



__Finally Graph Convoluitional Networks__
$$H^{(k+1)} = \sigma(AH^{(k)}W) \\ \Downarrow
\\ H^{(k+1)} =\sigma((A+I)H^{(k)}W^{(k)}) \\ \Downarrow
\\ H^{(k+1)} =  \sigma((\tilde{D}^{-1 / 2}(A+I) \tilde{D}^{-1 / 2})H^{(k)}W^{(k)}) \\ \Downarrow \\
\therefore H^{(k+1)} = \sigma(\tilde{A}H^{(k)}W^{(k)})$$ 

+ For each layer, the feature matrix and weight matrix are multiplied to create the next feature matrix


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1,2,3,4,5,6])
G.add_edges_from([(1,2),(1,3),(2,3),(1,4),(4,5),(4,6),(5,6)])

nx.draw(G, with_labels=True,node_size=600, font_size=22)
plt.show()
```

```python
A = nx.adjacency_matrix(G).todense()
print(A)
```

Assign feature vector $H$, so that it can be separated into two groups

```python
H = np.matrix([1,0,0,-1,0,0]).T
print(H)
```

Product of adgacency matrix and node features matrix represents the sum of neighboring node features

```python
# Sum of neighboring node feature
A*H
```

```python
A_self = A + np.eye(6)

# Sum features of neiboring nodes and it self
A_self*H
```

Similar to data pre-processing for any neural networks operation, normalize the features to prevent numerical instabilities and vanishing/expoding gradients in order for the model to converage.

```python
D = np.array(A_self.sum(1)).flatten()
D = np.diag(D)

D_half_norm = fractional_matrix_power(D,-0.5)

A_self = np.asmatrix(A_self)
D_half_norm = np.asmatrix(D_half_norm)

A_half_norm = D_half_norm * A_self * D_half_norm

A_half_norm*H
```

Build 2-layer GCN using ReLU as the activation function
$$H^{(2)}=ReLU(\tilde{A}H^{(1)}W^{(1)})$$
$$H^{(3)}=ReLU(\tilde{A}H^{(2)}W^{(2)})$$

```python
np.random.seed(20)

W1 = np.random.randn(1,4) #input: 1 -> hidden: 4
W2 = np.random.randn(4,2) # hidden: 4 -> output: 2

def relu(x):
    return np.maximum(0,x)

def gcn(A_self,H,W):
    D = np.diag(np.array(A_self.sum(1)).flatten())
    D_half_norm = fractional_matrix_power(D,-0.5)
    H_new = D_half_norm * A_self * D_half_norm * H * W
    return relu(H_new)

H1 = H
H2 = gcn(A_self, H1, W1)
H3 = gcn(A_self, H2, W2)

print(H3)
```
